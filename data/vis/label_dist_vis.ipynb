```python
def torch_load(base_dir, filename):
    fpath = os.path.join(base_dir, filename)    
    return torch.load(fpath, map_location=torch.device('cpu'))
```

```python
def wandb_save_boxplot(x, n_clients, dset, title , key):
    fig, ax = plt.subplots()

    ax.boxplot(x.values(), vert = False)
    for client_id in range(n_clients):
        ds = f'{client_id}-{dset}'
        
        ax.scatter(x = x[ds], y=np.random.normal(loc = client_id+1, scale= 0.15, size = x[ds].shape))    
    ax.set_title(title)
    wandb.log({key : wandb.Image(plt)})
```

## Cora Dataset

### Label Distribution over Clients via Boxplots
#### 5 Clients

```python
n_clients = 5
label_dict_tr_cora5, label_dict_val_cora5, label_dict_tst_cora5 =  {}, {}, {}
```

```python
for client_id in range(n_clients):
        ds = f'{client_id}-Cora'
        partition = torch_load(datapath, f'Cora_disjoint/{n_clients}/partition_{client_id}.pt')
        #Data objects
        tr, val, tst = partition['client_tr'], partition['client_val'] , partition['client_tst']
        label_dict_tr_cora5[ds] = tr.y.numpy()
        label_dict_val_cora5[ds] = val.y.numpy()
        label_dict_tst_cora5[ds] = tst.y.numpy()
```

```python
t1 = 'Training Label BoxPlot of Clients in Disjoint Cora-5'
t2 = 'Validation Label BoxPlot of Clients in Disjoint Cora-5'
t3 = 'Test Label BoxPlot of Clients in Disjoint Cora-5'
wandb_save_boxplot(label_dict_tr_cora5, n_clients, 'Cora', t1 , 'cora5-tr-label-boxplot')
wandb_save_boxplot(label_dict_val_cora5, n_clients, 'Cora', t2 , 'cora5-val-label-boxplot')
wandb_save_boxplot(label_dict_tst_cora5, n_clients, 'Cora', t3 , 'cora5-tst-label-boxplot')
```
